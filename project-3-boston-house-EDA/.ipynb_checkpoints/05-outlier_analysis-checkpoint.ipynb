```r
boston <- read.table("https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data", 
                     header = FALSE, sep = "")
names(boston) <- c("CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE","DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT",
                   "MEDV")
```

```r

boston$ZN <- boston$ZN + 1
boston$CHAS <- boston$CHAS + 1
boston_log = log(boston)
boston_log_sc = data.frame(scale(boston_log))
```

```r

display_outliers <- function (dataframe, feature, param=1.5) {
    feature_vec =  as.vector(dataframe[[feature]])
    Q1 <- quantile(feature_vec, .25)
    Q3 <- quantile(feature_vec, .75)
    tukey_window <- param*(Q3-Q1)
    less_than_Q1 <- dataframe[[feature]] < Q1 - tukey_window
    greater_than_Q3 <- dataframe[[feature]] > Q3 + tukey_window
    tukey_mask <- (less_than_Q1 | greater_than_Q3)
    return(dataframe[tukey_mask,])
}
```

```r
display_outliers(boston_log_sc, 'ZN')
display_outliers(boston_log_sc, 'CHAS')
```

```r
for (feature in colnames(boston_log_sc)){
    outlier_count = dim(display_outliers(boston_log_sc, feature))[1]
    print(paste(feature, outlier_count))
}
```

```r
raw_outliers = c()
for (feature in colnames(boston_log_sc)){
    outlier_df = display_outliers(boston_log_sc, feature)
    outlier_indices = rownames(outlier_df)
    raw_outliers = c(raw_outliers, outlier_indices)
}
raw_outliers
```
